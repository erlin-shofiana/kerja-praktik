```python
median_value = data["nonesensial_kapita"].median()
data["nonesensial_kapita"].fillna(median_value, inplace=True)
```

```python
data['nonesens_10persen'] = data.apply(
    lambda row: 0.1 * row['nonesensial_kapita'] if pd.isna(row['nonesens_10persen']) else row['nonesens_10persen'], 
    axis=1
)
```

```python
data['nonesens_15persen'] = data.apply(
    lambda row: 0.15 * row['nonesensial_kapita'] if pd.isna(row['nonesens_10persen']) else row['nonesens_10persen'], 
    axis=1
)
```

```python
data["umurxbobot"] = data["umur"]*data["weind"]
data["nonesensial_kapitaxbobot"] = data["nonesensial_kapita"]*data["weind"]
```

```python
data2019 = data[data["tahun"]==2019]
data2020 = data[data["tahun"]==2020]
data2021 = data[data["tahun"]==2021]
data2022 = data[data["tahun"]==2022]
data2023 = data[data["tahun"]==2023]
```

```python
result_bobot_2019 = data2019.groupby('Provinsi')['weind'].sum().reset_index()
result_bobot_2020 = data2020.groupby('Provinsi')['weind'].sum().reset_index()
result_bobot_2021 = data2021.groupby('Provinsi')['weind'].sum().reset_index()
result_bobot_2022 = data2022.groupby('Provinsi')['weind'].sum().reset_index()
result_bobot_2023 = data2023.groupby('Provinsi')['weind'].sum().reset_index()
combined_bobot = pd.concat([result_bobot_2019, result_bobot_2020, result_bobot_2021, result_bobot_2022, result_bobot_2023])
tahun_values = [2019] * 34 + [2020] * 34 + [2021] * 34 + [2022] * 34 + [2023] * 34
combined_bobot['Tahun'] = tahun_values[:len(combined_bobot)]
combined_bobot
```

```python
## Umur #
result_umur_2019 = data2019.groupby('Provinsi')['umurxbobot'].sum().reset_index()
result_umur_2020 = data2020.groupby('Provinsi')['umurxbobot'].sum().reset_index()
result_umur_2021 = data2021.groupby('Provinsi')['umurxbobot'].sum().reset_index()
result_umur_2022 = data2022.groupby('Provinsi')['umurxbobot'].sum().reset_index()
result_umur_2023 = data2023.groupby('Provinsi')['umurxbobot'].sum().reset_index()


combined_umur = pd.concat([result_umur_2019, result_umur_2020, result_umur_2021, result_umur_2022, result_umur_2023])
tahun_values = [2019] * 34 + [2020] * 34 + [2021] * 34 + [2022] * 34 + [2023] * 34
combined_umur['Tahun'] = tahun_values[:len(combined_umur)]
combined_umur
```

```python
## Non Esensial Kapita #
result_nonesensial_2019 = data2019.groupby('Provinsi')['nonesensial_kapitaxbobot'].sum().reset_index()
result_nonesensial_2020 = data2020.groupby('Provinsi')['nonesensial_kapitaxbobot'].sum().reset_index()
result_nonesensial_2021 = data2021.groupby('Provinsi')['nonesensial_kapitaxbobot'].sum().reset_index()
result_nonesensial_2022 = data2022.groupby('Provinsi')['nonesensial_kapitaxbobot'].sum().reset_index()
result_nonesensial_2023 = data2023.groupby('Provinsi')['nonesensial_kapitaxbobot'].sum().reset_index()


combined_nonesensial = pd.concat([result_nonesensial_2019, result_nonesensial_2020, result_nonesensial_2021, result_nonesensial_2022, result_nonesensial_2023])
tahun_values = [2019] * 34 + [2020] * 34 + [2021] * 34 + [2022] * 34 + [2023] * 34
combined_nonesensial['Tahun'] = tahun_values[:len(combined_nonesensial)]
combined_nonesensial
```

```python
## GK ##
result_gk_2019 = data2019.groupby('Provinsi')['gk'].mean().reset_index()
result_gk_2020 = data2020.groupby('Provinsi')['gk'].mean().reset_index()
result_gk_2021 = data2021.groupby('Provinsi')['gk'].mean().reset_index()
result_gk_2022 = data2022.groupby('Provinsi')['gk'].mean().reset_index()
result_gk_2023 = data2023.groupby('Provinsi')['gk'].mean().reset_index()
combined_gk = pd.concat([result_gk_2019, result_gk_2020, result_gk_2021, result_gk_2022, result_gk_2023])
tahun_values = [2019] * 34 + [2020] * 34 + [2021] * 34 + [2022] * 34 + [2023] * 34
combined_gk['Tahun'] = tahun_values[:len(combined_gk)]
combined_gk
```

```python
## Persentase Laki Laki #
data2019laki = data2019[data2019["gender"]=="Laki-Laki"]
result_laki_2019 = data2019laki.groupby('Provinsi')['weind'].sum().reset_index()

data2020laki = data2020[data2020["gender"]=="Laki-Laki"]
result_laki_2020 = data2020laki.groupby('Provinsi')['weind'].sum().reset_index()

data2021laki = data2021[data2021["gender"]=="Laki-Laki"]
result_laki_2021 = data2021laki.groupby('Provinsi')['weind'].sum().reset_index()

data2022laki = data2022[data2022["gender"]=="Laki-Laki"]
result_laki_2022 = data2022laki.groupby('Provinsi')['weind'].sum().reset_index()

data2023laki = data2023[data2023["gender"]=="Laki-Laki"]
result_laki_2023 = data2023laki.groupby('Provinsi')['weind'].sum().reset_index()

combined_laki = pd.concat([result_laki_2019, result_laki_2020, result_laki_2021, result_laki_2022, result_laki_2023])
tahun_values = [2019] * 34 + [2020] * 34 + [2021] * 34 + [2022] * 34 + [2023] * 34
combined_laki['Tahun'] = tahun_values[:len(combined_laki)]
combined_laki
```

```python
## Persentase Rural #
data2019rural = data2019[data2019["residence"]=="Rural"]
result_rural_2019 = data2019rural.groupby('Provinsi')['weind'].sum().reset_index()

data2020rural = data2020[data2020["residence"]=="Rural"]
result_rural_2020 = data2020rural.groupby('Provinsi')['weind'].sum().reset_index()

data2021rural = data2021[data2021["residence"]=="Rural"]
result_rural_2021 = data2021rural.groupby('Provinsi')['weind'].sum().reset_index()

data2022rural = data2022[data2022["residence"]=="Rural"]
result_rural_2022 = data2022rural.groupby('Provinsi')['weind'].sum().reset_index()

data2023rural = data2023[data2023["residence"]=="Rural"]
result_rural_2023 = data2023rural.groupby('Provinsi')['weind'].sum().reset_index()

combined_rural = pd.concat([result_rural_2019, result_rural_2020, result_rural_2021, result_rural_2022, result_rural_2023])
tahun_values = [2019] * 33 + [2020] * 33 + [2021] * 33 + [2022] * 33 + [2023] * 33
combined_rural['Tahun'] = tahun_values[:len(combined_rural)]

# Create DataFrame for DKI Jakarta with weind set to 0 for each year
dki_jakarta_rural = pd.DataFrame({
    'Provinsi': ['Dki Jakarta'] * 5,
    'weind': [0] * 5,
    'Tahun': [2019, 2020, 2021, 2022, 2023]
})

# Concatenate with the existing combined_rural DataFrame
combined_rural = pd.concat([combined_rural, dki_jakarta_rural], ignore_index=True)
combined_rural = combined_rural.sort_values(by=['Tahun', 'Provinsi']).reset_index(drop=True)
combined_rural
```

```python
new_index = [i % 34 for i in range(len(combined_rural))]

# Set the new index to the DataFrame
combined_rural['New_Index'] = new_index

# Optionally, if you want to set this as the DataFrame's index
combined_rural.set_index('New_Index', inplace=True)
combined_rural
```

## Data Fix

```python
import pandas as pd

# Assuming combined_* are existing DataFrames or Series
datafix = pd.DataFrame()

# Assign columns to datafix from combined_* DataFrames/Series
datafix["Provinsi"] = combined_bobot["Provinsi"]
datafix["Tahun"] = combined_bobot["Tahun"]
datafix["PersenUnmetNeed"] = combined_unmet["weind"] / combined_bobot["weind"]
datafix["PersenRural"] = combined_rural["weind"] / combined_bobot["weind"]
datafix["PersenTidakBekerja"] = combined_nobekerja["weind"] / combined_bobot["weind"]
datafix["PersenLakiLaki"] = combined_nobekerja["weind"] / combined_bobot["weind"]
datafix["PersenNoCoverage"] = combined_nocoverage["weind"] / combined_bobot["weind"]
datafix["Umur"] = combined_umur["umurxbobot"] / combined_bobot["weind"]
datafix["nonesensial_kapita"] = combined_nonesensial["nonesensial_kapitaxbobot"] / combined_bobot["weind"]
datafix["gk"] = combined_gk["gk"]
datafix["PersenKelas3_15"] = combined_kelas3_15["weind"] / combined_bobot["weind"]

# Display the resulting DataFrame
datafix
```
